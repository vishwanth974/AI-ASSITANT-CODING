Task Description #1
• Zero-shot: Prompt AI to write a function that checks whether a given year is a leap
year

Prompt : write a function that checks whether a given year is a leap year.

```python
def is_leap_year(year):
  """Checks if a given year is a leap year."""
  return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Get year from user input
try:
  year = int(input("Enter a year to check: "))
  if is_leap_year(year):
    print(f"{year} is a leap year.")
  else:
    print(f"{year} is not a leap year.")
except ValueError:
  print("Invalid input. Please enter a valid year.")
```

converts centimeters to inches.


writing an python function that converts centimeters to inches which takes input from user example: 5 cm = 1.9685

```python
def cm_to_inches(cm):
  """Converts centimeters to inches."""
  return cm * 0.393701

# Get centimeters from user input
try:
  cm = float(input("Enter a value in centimeters to convert to inches: "))
  inches = cm_to_inches(cm)
  print(f"{cm} cm is equal to {inches} inches.")
except ValueError:
  print("Invalid input. Please enter a valid number for centimeters.")
```

ask Description #3 :Few-shot: Provide 2–3 examples to generate a function that formats full names as
“Last, First”.


write a python function that formats full names as “Last, First” example : Input: John Smith → Output: Smith, John Input: Alice Johnson → Output: Johnson, Alice, take the input from user and label what is first name and what is last name

```python
def format_name_with_labels(full_name):
  """Formats a full name as 'Last, First' and labels the names."""
  names = full_name.split()
  if len(names) > 1:
    last_name = names[-1]
    first_name = ' '.join(names[:-1])
    return f"Last Name: {last_name}, First Name: {first_name}"
  else:
    return f"Full Name: {full_name}" # Return the name as is if only one name is provided and label it as Full Name

# Examples
print(format_name_with_labels("John Smith"))
print(format_name_with_labels("Alice Johnson"))

# Example with user input
try:
  name = input("Enter a full name to format: ")
  formatted_name = format_name_with_labels(name)
  print(f"Formatted name: {formatted_name}")
except ValueError:
  print("Invalid input. Please enter a valid name.")
```

Task Description#4
• Compare zero-shot and few-shot prompts for writing a function that counts the
number of vowels in a string.

Zero-Shot Prompt:Write a function that takes a string input from the user and returns the number of vowels in it.

```python
def count_vowels(text):
  """Counts the number of vowels in a string."""
  vowels = "aeiouAEIOU"
  vowel_count = 0
  for char in text:
    if char in vowels:
      vowel_count += 1
  return vowel_count

# Get string input from user
text_input = input("Enter a string: ")
num_vowels = count_vowels(text_input)
print(f"The number of vowels in the string is: {num_vowels}")
```

Few-Shot Prompt :Write a function that takes a string input from the user and returns the number of vowels in it. example: Input: "hello" → Output: 2 Input: "world" → Output: 1

```python
def count_vowels(text):
  """Counts the number of vowels in a string."""
  vowels = "aeiouAEIOU"
  vowel_count = 0
  for char in text:
    if char in vowels:
      vowel_count += 1
  return vowel_count

# Few-shot examples:
# Input: "hello" -> Output: 2
# Input: "world" -> Output: 1

# Get string input from user
text_input = input("Enter a string: ")
num_vowels = count_vowels(text_input)
print(f"The number of vowels in the string is: {num_vowels}")
```

Task Description#5
• Use few-shot prompting to generate a function that reads a .txt file and returns the
number of lines.

Write a function that takes a .txt filename as input from the user and returns the number of lines in the file.

```python
def count_lines_in_file(filename):
  """Counts the number of lines in a text file."""
  try:
    with open(filename, 'r') as f:
      line_count = sum(1 for line in f)
    return line_count
  except FileNotFoundError:
    return f"Error: File '{filename}' not found."
  except Exception as e:
    return f"An error occurred: {e}"

# Create a sample text file for demonstration
sample_filename = "sample.txt"
with open(sample_filename, "w") as f:
  f.write("This is the first line.\n")
  f.write("This is the second line.\n")
  f.write("This is the third line.\n")
  f.write("This is the fourth line.\n")

print(f"Created a sample file named '{sample_filename}'")

# Get filename from user input
filename_input = input("Enter the name of the .txt file (e.g., sample.txt): ")
num_lines = count_lines_in_file(filename_input)
print(f"The number of lines in '{filename_input}' is: {num_lines}")
```
